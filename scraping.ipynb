```python
def getAdditionalDetails(url):
    try:
        country_page = getHTMLContent('https://en.wikipedia.org/wiki/' + url)
        # print(country_page)
        table = country_page.find('table', {'class': 'infobox ib-country vcard'})
        additional_details = []
        read_content = False
        for tr in table.find_all('tr'):
            if (tr.get('class') == ['mergedtoprow'] and not read_content):
                link = tr.find('a')
                if (link and (link.get_text().strip() == 'Area' or
                   (link.get_text().strip() == 'GDP' and tr.find('span').get_text().strip() == '(nominal)'))):
                    read_content = True
            elif ((tr.get('class') == ['mergedrow'] or tr.get('class') == ['mergedbottomrow']) and read_content):
                additional_details.append(tr.find('td').get_text().strip('\n'))
                if (tr.find('div').get_text().strip() != '•\xa0Total area' and
                   tr.find('div').get_text().strip() != '•\xa0Total'):
                    read_content = False
                    print("IF:",tr.find('div').get_text().strip())
                else:
                    print("ELSE:",tr.find('div').get_text().strip(),read_content)
        return additional_details
    except Exception as error:
        print('Error occured: {}'.format(error))
        return []
```

```python
data_content = []
for row in rows:
    cells = row.find_all('td')
    if len(cells) > 1:
        try:
            country_link = cells[0].find('a')
            country_info = [cell.text.strip('\n') for cell in cells]
            print(country_info)
            if country_link.get('href').find('Demographics'):
                additional_details = getAdditionalDetails(country_link.get('href')[22:])

            else:
                additional_details = getAdditionalDetails(country_link.get('href'))

            if (len(additional_details) == 4):
                country_info += additional_details
                data_content.append(country_info)
        except:
            pass


dataset = pd.DataFrame(data_content)
```

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url="https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
```

```python
html_response=requests.get(url)
```

```python
soup=BeautifulSoup(html_response.text,'html.parser')
```

```python
population_table=soup.find('table',{'class':'wikitable'})
```

```python
#extract data
data_list=[]
```

```python
for row in population_table.find_all('tr')[1:]:
    columns=row.find_all(['th','td'])
    country=columns[1].text.strip()
    population=columns[2].text.strip()
    data_list.append([country,population])
```

```python
#open
#write mode
#with
file="population_data_retrieved.csv"
```

```python
with open(file,'w',newline='',encoding='utf-8') as csv_file:
    c=csv.writer(csv_file)
    c.writerow(['country','population']) #header row
    c.writerows(data_list)
```

```python
#load the csv file into a pandas dataframe
df=pd.read_csv('student_grades.csv')
#calculate and display the average grade for each subject
avg_grades=df[['Math','English','Science']].mean()
print("Average grades")
print(avg_grades)
#Identify and display the students with highest avg grade
highest_avg_student=df.loc[(df[['Math','English','Science']].mean(axis=1)==df[['Math','English','Science']].mean().max())]
print("student with Highest Average:")
print(highest_avg_student)
#create a new column called Total representing the total grade
df['Total']=df[['Math','English','Science']].sum(axis=1)
#Save the updated dataframe to a new csv file
df.to_csv('new_data.csv',index=False)
```
