#Amazon Product Reviews

```python
link = 'https://www.amazon.in/New-Apple-iPhone-12-64GB/product-reviews/B08L5TGWD1?reviewerType=all_reviews'
page = requests.get(link)
```

```python
soup = bs(page.content)
```

```python
print(soup.prettify)
```

```python
#Reviewer or customer names
names = soup.find_all('span' , class_ = 'a-profile-name')
```

```python
cust_name = []
for name in names:
  cust_name.append(name.get_text())
```

```python
print(cust_name)
```

```python
cust_name.pop(0)
cust_name.pop(1)
```

```python
print(cust_name)
```

```python
print(len(cust_name))
```

```python
#Review Titles
review_titles = soup.find_all('a',{"data-hook":"review-title"})
```

```python
review_headers = []
for rvt in review_titles:
  review_headers.append(rvt.get_text().strip())
```

```python
print(review_headers)
```

```python
print(len(review_headers))
```

```python
#Star Ratings
ratings = soup.find_all('i',{"data-hook":"review-star-rating"})
ratings
```

```python
star_ratings = []
for rates in ratings:
  star_ratings.append(rates.get_text())
```

```python
print(star_ratings)
print(len(star_ratings))
```

```python
reviews = soup.find_all("span",{"data-hook":"review-body"})
```

```python
product_reviews = []
for rev in reviews:
  product_reviews.append(rev.get_text().strip('\n').strip())
```

```python
print(product_reviews)
print(len(product_reviews))
```

```python
#cust_name
#review_headers
#star_ratings
#product_reviews
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
df['Customer Name'] = cust_name
df['Review Header'] = review_headers
df['Star Ratings'] = star_ratings
df['Product Reviews'] = product_reviews
```

```python
df
```

```python
df.to_csv('I-Phone_11_review.csv')
```
