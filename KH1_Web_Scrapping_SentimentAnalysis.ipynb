```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://www.amazon.in/product-reviews/B089MVKH71/'
```

```python
page = requests.get(link)
```

```python
page
```

```python
soup = bs(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
names = soup.find_all('span' , class_ = 'a-profile-name')
names
```

```python
cust_name = []
for name in names:
  cust_name.append(name.get_text())
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(0)
```

```python
cust_name
```

```python
title = soup.find_all('a' , class_='review-title-content')
title
```

```python
review_title = []
for t in title:
  review_title.append(t.get_text().strip('\n'))
```

```python
review_title
```

```python
import pandas as pd
df = pd.DataFrame()
```

```python
df['name'] = cust_name
df['review_title'] = review_title
```

```python
df
```

```python
rating = soup.find_all('i' , class_='review-rating')
rating
```

```python
rate = []
for r in rating:
  rate.append(r.get_text().strip('\n'))
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
df['ratings'] = rate
```

```python
df
```

```python
review = soup.find_all('span' , class_='review-text-content')
review
```

```python
reviews = []
for r in review:
  reviews.append(r.get_text().strip('\n'))

len(reviews)
```

```python
df['reviews'] = reviews
```

```python
df
```

```python
from textblob import TextBlob
```

```python
review_sentiments = []
for rev in df.reviews:
  review_sentiments.append(TextBlob(rev).sentiment.polarity)
```

```python
df['sentiments'] = review_sentiments
```
